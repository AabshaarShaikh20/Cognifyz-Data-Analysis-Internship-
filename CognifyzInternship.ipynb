# Task2: Cuisine Combination
1. Identify the most common combinations of cuisines in the dataset.
2. Determine if certain cuisine combinations tend to have higher ratings.

```python
# Task 2.1 -  Identify the most common combinations of cuisines

from collections import Counter

#Since the cuisine combinations are stored as tuples, we need to convert them into strings for visualization.
# So spliting the 'Cuisines' into combinations and thencounting the most common ones
cuisine_combinations = df['Cuisines'].str.split(', ').apply(tuple)
most_common_combinations = Counter(cuisine_combinations).most_common(10)

# Separating the combinations and their counts for plotting
combinations, counts = zip(*most_common_combinations)

# Converting tuples of cuisine combinations into strings
combinations_str = [' & '.join(cuisine) for cuisine in combinations]

# Here we will use a color map to generate different colors automatically instead of using single color color or specifying each color manually
colors = plt.cm.Paired(range(len(combinations_str)))

# Ploting the top 10 most common cuisine combinations
plt.figure(figsize=(10, 6))
plt.barh(combinations_str, counts, color=colors)
plt.xlabel('Number of Restaurants')
plt.title('Top 10 Most Common Cuisine Combinations')
plt.show()
```

- plt.cm.Paired() is a color map that provides a set of distinct colors. You can use it to generate a range of colors based on the number of bars.

- range(len(combinations_str)) ensures that each bar gets a different color from the color map automatically.

- combinations_str converts each tuple of cuisine combinations into a single string using ' & '.join(cuisine). This makes the data compatible for plotting.

```python
# Another way of doing it
common_cuisine_combinations = df['Cuisines'].value_counts().head(5)
print("Most common cuisine combinations:\n", common_cuisine_combinations)
```

1. Plot the locations of restaurants on a map using longitude and latitude coordinates.
2. Identify any patterns or clusters of restaurants in specific areas.

```python
# Task 3.1 - Plot the locations of restaurants on a map using longitude and latitude coordinates.
import cartopy.crs as ccrs
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
plt.scatter(df['Longitude'], df['Latitude'], color='red', marker='o', s=10, transform=ccrs.PlateCarree())
ax.coastlines()
plt.title('Locations of Restaurants')
plt.show()
```

```python
# Task 3.2 - Identify any patterns or clusters of restaurants in specific areas.
from sklearn.cluster import KMeans
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
X = df[['Latitude', 'Longitude']]

#using KMeans to find clusters
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['cluster'], cmap='viridis', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Restaurants')
plt.colorbar(label='Cluster')
plt.show()
```

# Task4: Restaurant Chains
1. Identify if there are any restaurant chains present in the dataset.
2. Analyze the ratings and popularity of different restaurant chains.

```python
# Task 4.1: Identify if there are any restaurant chains
restaurant_chains = df['Restaurant Name'].value_counts().loc[lambda x: x > 1]
print("Restaurant chains - \n" , restaurant_chains)
```

```python
# Task 4.2: Analyze the ratings and popularity of different restaurant chains
restaurant_names = df['Restaurant Name']
restaurant_name_counts = restaurant_names.value_counts()
chains = restaurant_name_counts[restaurant_name_counts > 1].index.tolist()
chain_dataframes = []
for chain in chains:
    chain_data = df[df['Restaurant Name'] == chain]
    average_rating = chain_data['Aggregate rating'].mean()
    total_votes = chain_data['Votes'].sum()
    chain_metrics = pd.DataFrame({'Chain': [chain], 'Average Rating': [average_rating], 'Total Votes': [total_votes]})
    chain_dataframes.append(chain_metrics)
chain_metrics = pd.concat(chain_dataframes, ignore_index=True)
chain_metrics.sort_values(by='Average Rating', ascending=False, inplace=True)
print("Chain-level metrics:")
print(chain_metrics.head(10))
```

1. Analyze the text reviews to identify the most common positive and negative keywords.
2. Calculate the average length of reviews and explore if there is a relationship between review length and rating.

```python
# Task 1.1: Analyze the text reviews to identify the most common positive and negative keywords

text_review = df['Rating text'].value_counts()
text_review.reset_index()
```

```python
# ploting the text reviews on the 
plt.figure(figsize=(7, 5)) 
plt.pie(text_review.values, labels = text_review.index, autopct = "%.1f%%")
plt.show()
```

```python
# Task 1.2: Calculate the average length of reviews and explore if there is a relationship between review length and rating.


# Calculating average review length:
reviews = df['Rating text']
total_review = len(reviews)
avg_review_length = sum(len(review.split()) for review in reviews) / total_review
avg_review_length = round(avg_review_length, 2)
print('Average Review Length:', avg_review_length)
```

```python
# analysing the relationship between ratings and review length:
review_len_vs_rating = {}

for rating in df['Aggregate rating'].unique():
    filtered_reviews = df.loc[df['Aggregate rating'] == rating]['Rating text']
    total_length = 0
    total_reviews = len(filtered_reviews)
    
    for review in filtered_reviews:
        total_length += len(review.split())
    
    average_length = total_length / total_review
    review_len_vs_rating[rating] = average_length
    
print('Ratung v/s Average reviews length')
print(review_len_vs_rating)
```

```python
# Calculate the average review length for each rating and store it in a dictionary
review_len_vs_rating = df.groupby('Aggregate rating')['Rating text'].apply(lambda x: x.str.split().str.len().mean()).to_dict()
print('Rating v/s Average reviews length')
print(review_len_vs_rating)
```

```python
# Create a DataFrame from the review_len_vs_rating dictionary
df_review_len_vs_rating = pd.DataFrame(list(review_len_vs_rating.items()), columns=['Rating', 'Average Review Length'])
df_review_len_vs_rating
```

```python
# Calculate the correlation between review length and rating
review_lengths = df['Rating text'].apply(lambda x: len(x.split()))
correlation = review_lengths.corr(df['Aggregate rating'])

print('Correlation between review length and rating:', correlation)
```

- With review length calculations we observed that there is negative reltion between review length and rating, meaning that as the length of the review increases, the aggregate rating tends to decrease.

# Task3: Price Range vs. Online Delivery and Table Booking
1. Analyze if there is a relationship between the price range and the availability of online delivery and table booking.
2. Determine if higher-priced restaurants are more likely to offer these services.

```python
# Task 3.1: Analyze if there is a relationship between the price range and the availability of online delivery and table booking
price_online_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'])
price_table_booking = pd.crosstab(df['Price range'], df['Has Table booking'])

# Plot Price v/s Online Delivery
plt.figure(figsize=(6, 3))
plt.bar(price_online_delivery.index, price_online_delivery['No'], label='No')
plt.bar(price_online_delivery.index, price_online_delivery['Yes'], bottom=price_online_delivery['No'], label='Yes')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Price Range v/s Online Delivery')
plt.legend()
plt.show()

# Plot Price v/s Table Booking
plt.figure(figsize=(6, 3))
plt.bar(price_table_booking.index, price_table_booking['No'], label='No')
plt.bar(price_table_booking.index, price_table_booking['Yes'], bottom=price_table_booking['No'], label='Yes')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Price Range v/s Table Booking')
plt.legend()
plt.show()
```

```python
# Task 3.2: Determine if higher-priced restaurants are more likely to offer these services

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)

sns.countplot(x='Price range' , hue='Has Online delivery' , data=df,palette='viridis')
plt.title('Online Delivery Avalibility by Price Range')

plt.subplot(1,2,2)
sns.countplot(x='Price range', hue='Has Table booking', data=df,palette='viridis')
plt.title('Table Booking Availibility by Price range')

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# Creating a crosstabs for price range vs. online delivery and table booking
cross_tab_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'])
cross_tab_booking = pd.crosstab(df['Price range'], df['Has Table booking'])

# Performing Chi-square tests for both online delivery & table booking
chi2_delivery, p_value_delivery, _, _ = stats.chi2_contingency(cross_tab_delivery)
chi2_booking, p_value_booking, _, _ = stats.chi2_contingency(cross_tab_booking)

# Setting significance level (alpha)
alpha = 0.05  # Typically, we use 0.05

# Determine if higher-priced restaurants are more likely to offer these services
higher_priced_delivery = p_value_delivery < alpha
higher_priced_booking = p_value_booking < alpha

# Output
print('Are higher priced restaurants more likely to offer Online delivery?', higher_priced_delivery)
print('Are higher priced restaurants more likely to offer booking?', higher_priced_booking)
```
